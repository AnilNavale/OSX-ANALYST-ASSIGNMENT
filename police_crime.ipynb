```python
df["lat"] = df['location'].apply(lambda x : x.split(',')[0][1:])
```

```python
df["long"] = df['location'].apply(lambda x : x.split(',')[1][:-1])
```

```python
df["lat"] = df['lat'].apply(lambda x : float(x))
```

```python
df["long"] = df['long'].apply(lambda x : float(x))
```

```python
crime_vso_locations=df[df['category']=="ASSAULT"].groupby(['lat','long']).sum()['COUNTER'].reset_index()
crime_vso_locations.sort_values(by='COUNTER',ascending=False,inplace=True)
crime_vso_locations=crime_vso_locations.head(25)
```

```python
import plotly.express as px
fig=px.scatter_mapbox(crime_vso_locations, lat="lat", lon="long",color="COUNTER",size="COUNTER",
                      title="Top 25 locations with the highest volume ASSUALT",
                      color_continuous_scale=px.colors.sequential.thermal)
fig.update_layout(mapbox_style="carto-positron")
fig.show()
```

```python

crime_type_df = df['category'].value_counts(ascending=True)


fig=plt.figure(figsize=(18, 16))
plt.title("Frequency of Crimes Per Crime Type")
plt.xlabel("Frequency of Crimes")
plt.ylabel("Type of Crime")
plt.style.use('dark_background')
ax = crime_type_df.plot(kind='barh')
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
```

```python

crime_type_df = df['resolution'].value_counts(ascending=True)


fig=plt.figure(figsize=(14, 12))
plt.title("Frequency of resolution")
plt.xlabel("Frequency of resolution")
plt.ylabel("Type ofresolution")
plt.style.use('dark_background')
ax = crime_type_df.plot(kind='barh')
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
```
